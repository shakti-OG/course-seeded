```
%%writefile main.cpp
#include <iostream>
#include <vector>
#include <cmath>
#include <algorithm>
#include <numeric>
using namespace std;

pair <int,int> simplify(int p, int q){
  return {p/gcd(p,q), q/gcd(p,q)};
}

pair <int,int> add(pair <int, int> x, pair <int,int> y){
  int p = x.first*y.second+x.second*y.first;
  int q = x.second*y.second;
  pair <int,int> ans = simplify(p,q);
  return ans;
}

int main (){
  vector <pair<int,int>> v = {
    {1,2}, {2,3}, {3,4}, {4,5}
  };
  pair <int,int> ans = {0,2};
  for (int i=0; i<v.size(); i++){
    ans = add(ans, v[i]);
    cout << ans.first << "/" << ans.second << "\n";
  }
}
```
