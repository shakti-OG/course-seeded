```
%%writefile operations.cpp
#include <iostream>
#include <vector>
#include <cmath>
#include <algorithm>
#include <limits>
#include <numeric>
using namespace std;

pair <int,int> simplify(int p, int q){
  if (q == 0) return {0,0};
  return {p/gcd(p,q), q/gcd(p,q)};
}

pair <int,int> add(pair <int, int> x, pair <int,int> y){
  int p = x.first*y.second+x.second*y.first;
  int q = x.second*y.second;
  if (q == 0) return {0,0};
  return simplify(p,q);
}

int main (){
  double inf = numeric_limits<double>::infinity();
  double nan = numeric_limits<double>::quiet_NaN();
  int int_nan = numeric_limits<int>::quiet_NaN();
  cout << inf << " " << nan << "\n";
  cout << int_nan << " " << int_nan*0 << "\n";
  vector <pair<int,int>> v = {
    {1,2}, {2,3}, {3,4}, {4,5}
  };
  pair <int,int> ans = {0,1};
  for (int i=0; i<v.size(); i++){
    cout << "ans_sec: " << ans.second << " v_i: " << v[i].second << " ";
    ans = add(ans, v[i]);
    cout << "ans: " << ans.first << "/" << ans.second << "\n";
  }
}
```

```
%%writefile power_function.cpp

#include <iostream>
#include <vector>
#include <cmath>
#include <algorithm>
#include <limits>
#include <numeric>
using namespace std;

int square (int x){
  return x*x;
}

int cube(int x){
  return x*x*x;
}

int power (int x, int n){
  int ans = 1;
  for (int i=1; i<=n; i++) ans = ans*x;
  return ans;
}

int main (){
  cout << power(2,3) << " " << square(2) << " " << cube(3);
}
```
